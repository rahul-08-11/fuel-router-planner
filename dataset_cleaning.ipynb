```python
df = pd.read_csv("fuel-prices-for-be-assessment.csv", dtype=str)
```

```python
# describe dataset
df.describe()
```

```python
# Standardize column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(r"\s+", "_", regex=True)
)
```

```python
df
```

```python
# Strip whitespace from all string columns
df = df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
```

```python
# Title-case name & city; uppercase state

df["truckstop_name"] = df["truckstop_name"].str.title()
df["city"]           = df["city"].str.title()
df["state"]          = df["state"].str.upper()
```

```python
df
```

```python
#  Convert numeric columns
df["rack_id"]       = pd.to_numeric(df["rack_id"], errors="coerce")
df["retail_price"]  = pd.to_numeric(df["retail_price"], errors="coerce")
```

```python
df
```

```python
# Before dropping duplicates

print(f"length of the dataframe: {len(df)}")
```

```python
#  Drop duplicate rows
df = df.drop_duplicates()
```

```python
# After dropping duplicates

print(f"length of the dataframe: {len(df)}")
```

```python
df
```

```python
# Reset index
df = df.reset_index(drop=True)
```

```python
# Attaching longtitude and  latitude of 31k Cities - downloadable here https://simplemaps.com/data/us-cities
```

```python
uscities_df = pd.read_csv("uscities.csv")[['city', 'state_id','lat','lng']].rename(columns={"state_id": "state"})
```

```python
uscities_df = uscities_df.drop_duplicates(subset=["city", "state"], keep="first")
```

```python
uscities_df = uscities_df.drop_duplicates(subset=["city", "state"], keep="first")

# save only langitude and latitude of cities as separate csv
uscities_df.to_csv("state_city_coordinates.csv", index=False)
```

```python
df_merged = df.merge(
    uscities_df,
    on=["city", "state"],
    how="left"  # keeps all rows from df1, adds lat/lng where match found
)

print(df_merged.shape)
final_df = pd.DataFrame(df_merged)
```
